### 2. Double Linked Lists

```python
class DoubleLinkedList :
    class Node :
        __slot__ = ['e,','prev','next']
        
        def __init__(self, e=None, prev=None, next=None) :
            self.e = e
            self.prev = prev
            self.next = next
            
    def __init__(self) :
        self.head = self.Node("head")
        self.tail = self.Node("tail")
        self.size = 0
        self.head.next = self.tail
        self.tail.prev = self.head
        
    def isEmpty(self) :
        return self.size == 0
    
    def checkNode(self, N) :
        tmpNode = self.head
        while tmpNode != self.tail.next :
            if tmpNode == N :
                return True
            tmpNode = tmpNode.next
        return False
    
    def push_back(self, N) :
        if self.isEmpty() :
            self.head.next = N
            N.prev = self.head
        else :
            self.tail.next = N
            N.prev = self.tail
        self.tail = N  
        self.size += 1
    
    def push_front(self, N) :
        if self.isEmpty() :
            self.head.next = N
            N.prev = self.head
            self.tail = N
        else :
            front = self.head.next
            front.prev = N
            N.next = front
            N.prev = self.head
            self.head.next = N  
        self.size += 1
    
    def insert_between(self, N, a, b) :
        N.prev = a
        N.next = b
        a.next = N
        b.prev = N
     
    def move_after(self, N) : #한 칸 뒤로
        if not self.checkNode(N) :
            print "no exists the Node"
            return False
        nxt = N.next
        prv = N.prev
        
        prv.next = nxt
        nxt.prev = prv

        nxt.next.prev = N
        N.next = nxt.next
        
        N.prev = nxt
        nxt.next = N
        
    def move_before(self, N) : #한 칸 앞으로
        if not self.checkNode(N) :
            print "no exists the Node"
            return False
        nxt = N.next
        prv = N.prev
        
        nxt.prev = prv
        prv.next = nxt
        
        prv.prev.next = N
        N.prev = prv.prev
        
        N.next = prv
        prv.prev = N
    
    def pop_back(self) :
        if self.isEmpty() :
            print "List is Empty"
        else :
            self.tail = self.tail.prev
            del self.tail.next
            self.size += -1
            
    def pop_front(self) :
        if self.isEmpty() :
            print "List is Empty"
        else :
            tmpNode = self.head.next.next
            del self.head.next
            self.head.next = tmpNode
            tmpNode.prev = self.head
        
    def printList(self, N) : # N is startNode
        tmpNode = N
        while tmpNode != self.tail :
            print tmpNode.e,
            tmpNode = tmpNode.next
        print tmpNode.e
        
```

```python
DLL = DoubleLinkedList()

DLL.isEmpty()
Head = DLL.head

A = DLL.Node('a')
B = DLL.Node('b')
C = DLL.Node('c')
D = DLL.Node('d')
E = DLL.Node('e')
F = DLL.Node('f')
G = DLL.Node('g')

DLL.push_back(D)
DLL.push_back(E)
DLL.push_back(F)
DLL.printList(Head) #head d e f

            
DLL.push_front(B)
DLL.push_front(A)
DLL.printList(Head) #head a b d e f


DLL.printList(B) #b d e f


DLL.insert_between(C,B,D)
DLL.printList(Head) #head a b c d e f

DLL.move_after(C)
DLL.printList(Head) #head a b d c e f

DLL.move_before(G) #no exists the Node

DLL.move_before(C)
DLL.printList(Head) #head a b c d e f

DLL.pop_back()
DLL.printList(Head) #head a b c d e

DLL.pop_front()
DLL.printList(Head) #head b c d e

```
